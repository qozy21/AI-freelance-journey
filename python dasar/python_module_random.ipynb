#Python standard library : Module Random

Modul `random` di python digunakan untuk menghasilkan angka psudo-acak.

Berikut beberapa fungsi umum pada module `random` :

* `random()`: mengembalikan bilangan float acak antara 0,0(inklusif) dan 1,0(eksklusif)
* `randint(a, b):` Mengembalikan bilangan bulat acak N sedemikian sehingga a <= N <= b.

* `randrange(start, stop, step):` Mengembalikan elemen yang dipilih secara acak dari `range(start, stop, step)`.

* `uniform(a, b):` Mengembalikan bilangan float acak N sedemikian sehingga a <= N <= b atau b <= N <= a.

* `choice(seq):` Mengembalikan elemen acak dari urutan `seq` yang tidak kosong.

* `shuffle(seq)`: Mengacak elemen-elemen dari sebuah sequence seq di tempatnya.
* `sample(population, k)`: Mengembalikan daftar baru yang berisi k elemen unik yang dipilih dari sequence atau set populasi.


```python
import random

#angka integer random
print(random.randint(1, 80))

#angka float random
print(random.random())
print(random.uniform(1, 10))

#pilih satu item
buah = ["apel", "jeruk", "mangga", "semangka"]
print(random.choice(buah))

#acak urutan
random.shuffle(buah)
print(buah)
```
