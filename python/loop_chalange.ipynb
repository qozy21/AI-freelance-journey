**Tugas 1**: Hitung Mundur Sederhana (while loop)

Buatlah sebuah program Python yang menggunakan while loop untuk melakukan hitung mundur dari angka 10 ke 1, dan setelah mencapai 1, program mencetak "Luncurkan!".
Kriteria:
Gunakan while loop.
Mulai dari angka = 10.
Cetak setiap angka hitung mundur.
Cetak "Luncurkan!" setelah loop selesai.

```python
# Penghitung Mundur
masukan_angka = int(input("Masukkan angka: "))
angka = masukan_angka

while angka >= 1:
    print(angka)
    angka -= 1

print("Luncurkan!")
```

**Tugas 2**: Menghitung Total Belanja (for loop)

Anda memiliki daftar harga barang belanjaan. Gunakan for loop untuk menghitung total biaya belanjaan tersebut.
Kriteria:
Gunakan daftar harga yang sudah ditentukan dalam kode Anda: harga_barang = [15000, 25000, 10000, 5000, 50000]
Gunakan for loop untuk menjumlahkan semua elemen dalam daftar tersebut.
Cetak hasil total belanjaan di akhir.


```python
#Penghitung belanja

masukan_daftar_harga =input("Masukkan daftar harga(pisahkan dengan spasi): ").split( )
harga_barang = tuple(map(int, masukan_daftar_harga))
total_belanja = 0
print(harga_barang)

for harga in harga_barang:
    total_belanja += harga

print("Total belanja:", total_belanja)
```

**Tugas 3**: Menemukan Angka Genap (for loop dengan if)

Buatlah program yang mengiterasi angka dari 1 hingga 20 menggunakan for loop. Di dalam loop, gunakan pernyataan if untuk memeriksa apakah angka saat ini adalah angka genap, lalu cetak hanya angka genap tersebut.
Kriteria:
Gunakan for loop dengan range() dari 1 hingga 20.
Gunakan operator modulo (%) untuk mengecek genap (angka % 2 == 0).

```python
# pencetak angka genap
angka = int(input("Masukkan batas angka: "))
for i in range(1, 21):
    if i % 2 == 0:
        print(i)
```

**Tugas 4**: Validasi Password (while loop dengan break)

Buat program yang terus meminta pengguna memasukkan kata sandi (password) yang benar. Loop harus berhenti hanya ketika pengguna memasukkan kata sandi yang telah ditentukan sebelumnya.
Kriteria:
Tentukan password yang benar dalam variabel, misalnya PASSWORD_BENAR = "rahasia123".
Gunakan while True untuk membuat loop berjalan tanpa batas awalnya.
Gunakan input dari user di dalam loop.
Gunakan if untuk mencocokkan input dengan password yang benar.
Gunakan break untuk keluar dari loop jika password benar.

```python
# password validator
password_benar = "rahasia123"
kesempatan = 3
i=0
while i < kesempatan:
  masukan_password = input("Masukkan password: ")
  if masukan_password == password_benar:
    print("Password benar!")
    break
  i += 1
  if i == kesempatan:
    print("Kesempatan habis!")
```
