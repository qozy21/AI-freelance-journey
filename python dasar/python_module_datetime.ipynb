##**Python standard library: Module Datetime**

Modul `datetime` di Python menyediakan kelas untuk memanipulasi tanggal dan waktu dengan cara sederhana dan efektif. Modul ini memungkinkan untuk bekerja dengan tanggal, waktu, dan interval waktu.

Berikut beberapa kelas utama dalam module `datetime`:
* `datetime.date(2005, 12, 23)` : mengelela tanggal
* `datetime.time(15, 20, 0)` : mengelola waktu
* `datetime.datetime(2016, 10, 27, 15, 30, 0)`: menggabungkan tanggal dan watku
* `datetime.timedelta(days=7)` : mempresentasikan durasi atau perbedaan antara dua objek
* `tzinfo` : info zona waktu




```python
import datetime

#Tanggal dan waktu lengkap
print("sekarang:", datetime.datetime.now())

#Hanya Tanggal
print("Hari ini:", datetime.date.today())

#Hanya waktu
print("Sekarang:", datetime.datetime.now().time())

#Membuat tanggal tertentu
print("ulang tahun:", datetime.date(2016, 8, 22))

#Membuat waktu tertentu
print("waktu acara:", datetime.datetime(2016, 8, 22, 15, 30, 0))

#Format indonesia
print("Tanggal Indonesia:", datetime.datetime.now().strftime("%d/%m/%Y %H:%M:%S"))

#Format panjang
print("Tanggal panjang:", datetime.datetime.now().strftime("%A, %d %B %Y"))

#format kostum
print("Tanggal kostum:", datetime.datetime.now().strftime("%d-%m-%Y,jam %H: %M"))
```
