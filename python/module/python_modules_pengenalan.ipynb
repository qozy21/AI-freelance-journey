#Apa itu Modules ?
*  modules adalah file python(.py) yang berisi function, variable, dan statment yang bisa digunakan program lain

**Keuntungan memakai Modules**
* Code Reuse : tidak perlu menulis ulang function yang sama
* organization : membagi code menjadi file kecil yang mudah dikelola
* collaboration : tim bisa bekerja pada file yang berbeda
* Maintenance : mudah dalam perbakan
* Readability : code lebih mudah dibaca

**Docstring**

dokumentasi """teks""" (menggunakan petik 3) yang sering dipakai dalam fungsi,class,module. bisa beberapa baris sekaligus, dan bisa terbaca oleh __doc__. penempatannya pada baris pertama dalam suatu function

**\_\_name__ == "\_\_main__"**

akan menghasilkan true jika file dijalankan sebagai program utama bukan sebagai module

```python
%%writefile kalkulator.py
"""Modul kalkulator sederhana."""
# menulis file kalkulator.py sebagai module

def tambah(a, b):
    """Fungsi untuk menambahkan dua angka.""" #ini docstring
    return a + b

def kurang(a, b):
    """fungsi untuk mengurangi dua angka."""
    return a - b

def kali(a, b):
    """Fungsi untuk mengkalikan dua angka."""
    return a * b

def bagi(a, b):
    """Fungsi untuk membagi dua angka."""
    if b != 0:
        return a / b
    else:
        return "Tidak dapat membagi dengan nol"

if __name__ == "__main__": #jika file ini dijalankan sebagai program utama(bukan di import sebagai modul maka kode dibawahny akan dieksekusi)
    print("Ini adalah modul kalkulator.")
```

```python
#program utama
import kalkulator #import semua dari kalkulator

print("operasi 2 angka")
a = int(input("Masukkan angka pertama: "))
b = int(input("Masukkan angka kedua: "))

hasil_tambah = kalkulator.tambah(a, b)
hasil_kurang = kalkulator.kurang(a, b)
hasil_kali = kalkulator.kali(a, b)
hasil_bagi = kalkulator.bagi(a, b)

print("jika ditambahkan: " + str(hasil_tambah))
print("jika dikurangi: " + str(hasil_kurang))
print("jika dikalikan: " + str(hasil_kali))
print("jika dibagi: " + str(hasil_bagi))

print(kalkulator.__doc__)
print(kalkulator.tambah.__doc__)
```
