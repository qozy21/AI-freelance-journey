String methode adalah fungsi untuk mengolah data tipe string

```python
# upper(), lower(), capitalize(), title(), swapcase(), strip()
nama = "  saNdiaga Uno  "
print(nama.upper())       # "  SANDIAGA UNO  "
print(nama.lower())       # "  sandiaga uno  "
print(nama.capitalize())  # "  Sandiaga uno  "
print(nama.title())       # "  Sandiaga Uno  "
print(nama.swapcase())    # "  SAnDIAGA uNO  "
print(nama.strip())       # "sANDIAGA oNO"
```

# Operator

Operator adalah simbol khusus yang digunakan untuk melakukan operasi pada nilai atau variabel dan nilai. Beberapa jenis operator di python
* Operator Aritmatika
* Operator Assignments (penugasan)
* Operator Perbandingan
* Operator Logika
* Operator String



**Operator Aritmatik**
* operator dasar : + (tambah), - (kurang), * (kali), / (bagi)
* operator pangkat : ** (pangkat)
* operator pembagian bulat : // (pembagian bulat)
* operator modulo : % (modulo/sisabagi)

```python
a = 8
b = 3

# operator dasar
print(a + b)      # 11
print(a - b)      # 5
print(a * b)      # 24
print(a / b)      # 2.66666666

#operator pangkat
print(a ** b)     # 512

#operator pembagian bulat
print(a // b)     # 2 (hasil tanpa desimal)

#operator modulo
print(a % b)      # 2 (sisa pembagian)
```

**Operator Assigment**

untuk memberikan nilai pada varaiabel menggunakan (=). compound assigment adalah gabungan aritmatika dan assigment

```python
x = 10
print("x awal:", x)

x += 5      # x = x + 5
print("x setelah += 5:", x) # 15

x -= 3      # x = x - 3
print("x setelah -= 3:", x) # 12

x *= 2      # x = x * 2
print("x setelah *= 2:", x) # 24

x /= 4      # x = x / 4
print("x setelah /= 4:", x) # 6.0
```

**Operator Perbandingan**

akan menghasilkan True atau False

```python
a = 7
b = 4

print(a > b)  # True
print(a < b)  # False
print(a >= b) # True
print(a <= b) # False
print(a == b) # False (sama dengan)
print(a != b) # True (tidak sama dengan)

nama1 = 'Alice'
nama2 = 'Bob'
nama3 = 'Alice'

print(nama1 == nama2)  # False
print(nama1 != nama2)  # True
print(nama1 == nama3)  # True
```

**Operator Logika**

* and : true jika kedua kondisi true
* or : true jika salah satu kondisi true
* not : membalikan boolean

```python
a = 23
b = 24
print(a < b and b != a) # True
print(b > a or b == a)  # True

c = a < b
print(not c) # False
```

**Operator String**

* Concatenation (+), menambah string
* Repetition (*), mengulang string sejumlah angka
* Membership (in), mengecek apakah text ada dalam string


```python
nama_depan = "Fiqhi"
nama_tengah = "Haidar"
nama_belakang = "Qozvin"
nama_lengkap = nama_depan + " " + nama_tengah + " " + nama_belakang
print(nama_lengkap) # Fiqhi Haidar

salam = "hello "
print(salam * 3)  # hello hello hello

print("Haidar" in nama_lengkap) # True
print("andika" in nama_lengkap) # False
print("Haidar" not in nama_lengkap) # False
```

**Prioritas Operator**

Urutan Prioritas (dari tinggi ke rendah):

* ** (pangkat)
* *, /, //, % (perkalian, pembagian)
* +, - (penjumlahan, pengurangan)
* ==, !=, <, >, <=, >= (perbandingan)
* not (logika not)
* and (logika and)
* or (logika or)



# Kontrol Alur Program

adalah cara untuk mengatur jalannya program berdasarkan kondisi tertentu

**Pernyataan if**

* Pernyataan if digunakan untuk menjalankan kode hanya jika kondisi tertentu bernilai True.
* Setelah if, harus ada : (titik dua)
* Kode yang akan dijalankan jika kondisi True harus diindentasi (diberi spasi di awal)


```python
angka = int(input("masukan angka : "))

if angka % 2 == 0:
  print(f"{angka} adalah angka genap")

if angka % 2 != 0:
  print(f"{angka} adalah angka ganjil")

if angka > 0:
  print(f"{angka} adalah angka positif")

if angka < 0:
  print(f"{angka} adalah angka negatif")
```

**Pernyataan if-else**

if-else digunakan ketika kita ingin menjalankan kode yang berbeda untuk kondisi True dan False.


```python
nilai = int(input("masukan nilai : "))

if nilai >= 60:
  print("anda lulus")
else:
  print("anda tidak lulus")
```

**Pernyataan elif**

elif (else if) digunakan untuk mengecek beberapa kondisi secara berurutan.


```python
nilai = int(input("masukan nilai : "))

if nilai >= 90:
  print("A")
elif nilai >= 80:
  print("B")
elif nilai >= 70:
  print("C")
elif nilai >= 60:
  print("D")
else:
  print("E")
```

**Kondisi dengan operator logika**

kondisi harus bernilai boolean,sehingga bisa menggunakan operator, misal operator logika, seperti and, or dan not


```python
umur = int(input("masukan umur : "))
pendapatan = int(input("masukan pendapatan : "))

if umur >= 19 and pendapatan >= 25000:
  print("Anda adalah profesional muda yang sukses!")
else:
  print("Anda belum memenuhi kriteria")
```

**Nested If**

menempatkan if dalam if yang lain

```python
username = input("masukan username : ")
password = int(input("masukan password : "))

if username == "admin":
  if password == 12345:
    print("selamat datang admin")
  else:
    print("Password Salah")
else:
  print("Username tidak ditemukan")
```

**Pernyataan Match Case**

fitur yang lebih bersih dibandingkan elif yang banyak

```python
hari = input("masukan nama hari : ").lower().strip()

match hari:
  case "senin" | "selasa" | "rabu" | "kamis" | "jumat":
    print("Hari kerja")
  case "sabtu" | "minggu":
    print("Hari Libur")
  case _:
    print("Hari tidak valid")
```

**Conditional Expression (Ternary Operator)**

menulis kondisi sederhana dalam satu baris

```python
angka = int(input("masukan angka : "))

# dengan if-else biasa
if angka % 2 == 0:
  hasil = "genap"
else:
  hasil = "ganjil"

# dengan ternary operator
hasil = "genap" if angka % 2 == 0 else "ganjil"
print("Angka tersebut:", hasil)
```

# Perulangan (loop)

menjalankan kode yang sama berulang kali. macamnya :
* for loop
* for loop dengan string
* while loop
* break and continue
* loop dengan else

**For loop**

mengulang kode sejumlah tertentu, biasanya menggunakan function range()

```python
# for loop

for i in range(5):
  print("a", i)     # 0 sampai 4

for i in range(1,5):
  print("b", i)     # 1 sampai 4

for i in range(1,10,2):
  print("c", i)     # ganjil dari 1 sampai 9

for i in range(8,0,-1):
  print("d", i)     # mundur dari 8 sampai 1
```

**for loop dengan string**

digunakan untuk mengiterasi setiap kata dalam string

```python
# for loop dengan string

nama = "sandi"
for huruf in nama:
  print(huruf)

kata = input("masukan kata : ")
for karakter in kata:
  print("-",karakter)
```

**while loop**

mengulang kode selama kondisi masih true

```python
# while loop
countdown = int(input("hitung waktu mundur :"))

while countdown >= 0:
  print(countdown)
  countdown -= 1
```

**Break and continue**

* break digunakan untuk keluar dari perulangan
* continue digunakan untuk melanjutkan ke iterasi berikutnya


```python
# break
import random
angka_random = random.randint(1,10)

while True:
  tebakan = int(input("masukan angka: "))

  if tebakan == angka_random:
    print("jawaban benar")
    break
  else:
    print("coba lagi!")
```

```python
# continue
for i in range(10):
  if i % 2 == 0:
    continue
  print(i)    # 1 3 4 5 7 9
```

**Loop dengan else**

Python memiliki fitur unik dimana loop bisa memiliki clause else yang dieksekusi jika loop selesai secara normal (tidak dengan break).

```python
# for-else

kata = input("masukan kata: ")
huruf_dicari = input("Masukan huruf yang dicari: ")

for huruf in kata:
  if huruf == huruf_dicari:
    print("huruf", huruf_dicari, "ditemukan dalam kata!")
    break
else:
    print("huruf", huruf_dicari, "tidak ditemukan dalam kata!")
```

```python
# while else

password_benar = "belajar mandiri"
percobaan = 0
kesempatan = 3

while percobaan < kesempatan:
  password = input("masukan password: ")
  percobaan += 1

  if password == password_benar:
    print("Login berhasil")
    break
  else:
    print("password salah. sisa kesempatan:", kesempatan - percobaan)

else:
  print("maksimal kesempatan, akses ditolak")
```
