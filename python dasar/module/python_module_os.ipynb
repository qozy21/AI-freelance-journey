##**Python standard library : Module OS**

Modul OS adalah modul bawaan Python untuk berinteraksi dengan sistem operasi, seperti:
* Mengelola file dan folder
* Navigasi direktori
* menjalankan perintah sistem
* mendapatkan informasi environment
* DLL


```python
import os # Module untuk interaksi dengan OS
from pathlib import Path # Module untuk mengelola path
```

```python
# Informasi sistem dan environment

print(f"\n1. Nama OS:{os.uname()}")                     # nama sistem operasi nt untuk windows dan posix untuk linux/Mac

print(f"2. Direktori kerja saat ini: {os.getcwd()}")    # Direktori kerja saat ini

try:
  print(f"3. Username: {os.getlogin()}")                # Username
except:
  print(f"3. username: tidak ditemukan")

print(f"4. Informasi environment: {os.environ.get('HOME') or os.environ.get('USERPROFILE')}")       # Informasi environment

print(f"5.Path separator: {os.sep}")                                                                # Path separator sesuai os
```

```python
# Navigasi dan direktori

print(f"Direktori parent: {os.path.dirname(os.getcwd())}") # Direktori parent
```

### Apa itu Modul `os`?

Modul `os` (Operating System) adalah modul bawaan Python yang menyediakan cara untuk berinteraksi dengan fungsionalitas sistem operasi. Ini memungkinkan program Python Anda untuk melakukan tugas-tugas seperti:

*   Mengelola file dan folder (membuat, menghapus, mengubah nama, memeriksa keberadaan).
*   Navigasi direktori (mengubah direktori kerja, mendapatkan direktori saat ini).
*   Menjalankan perintah sistem operasi.
*   Mendapatkan informasi lingkungan sistem (variabel lingkungan, nama OS, dll.).
*   Bekerja dengan path (menggabungkan, memisahkan, mendapatkan nama file dari path).

### Kategori dan Contoh Penggunaan Modul `os`

#### 1. Informasi Sistem dan Lingkungan

Kategori ini mencakup fungsi-fungsi untuk mendapatkan informasi tentang sistem operasi, direktori kerja, dan variabel lingkungan.

```python
# Import module os
import os

# Informasi sistem dan environment

print(f"\n1. Nama OS:{os.uname()}")                     # nama sistem operasi (nt untuk Windows, posix untuk Linux/Mac)

print(f"2. Direktori kerja saat ini: {os.getcwd()}")    # Direktori kerja saat ini

try:
  print(f"3. Username: {os.getlogin()}")                # Username (mungkin tidak berfungsi di semua lingkungan, seperti Google Colab)
except:
  print(f"3. username: tidak ditemukan")

print(f"4. Informasi environment: {os.environ.get('HOME') or os.environ.get('USERPROFILE')}")       # Informasi environment (misalnya direktori home user)

print(f"5. Path separator: {os.sep}")                                                                # Path separator sesuai OS (misal: '/' di Linux/Mac, '\\' di Windows)
```

#### 2. Navigasi dan Manajemen Direktori

Fungsi-fungsi di kategori ini memungkinkan Anda untuk berpindah antar direktori dan mendapatkan informasi path.

```python
import os

print(f"Direktori parent dari direktori kerja saat ini: {os.path.dirname(os.getcwd())}")

# Contoh lain:
# os.chdir('/tmp') # Mengubah direktori kerja ke /tmp
# print(f"Direktori kerja setelah diubah: {os.getcwd()}")
# os.mkdir('new_folder') # Membuat folder baru
# print(f"Isi direktori: {os.listdir('.')}") # Melihat isi direktori saat ini
```

#### 3. Manajemen File

Modul `os` juga menyediakan fungsi untuk membuat, menghapus, menyalin, dan memindahkan file. Meskipun `shutil` sering digunakan untuk operasi file yang lebih kompleks, `os` menyediakan dasar-dasarnya.

```python
import os

# Membuat file dummy
with open('my_file.txt', 'w') as f:
    f.write('Hello, OS module!')
print(f"File 'my_file.txt' dibuat: {os.path.exists('my_file.txt')}")

# cek apakah direktori/file ada
print("file ada:", os.path.exist("my_file.txt"))

# Mengganti nama file
os.rename('my_file.txt', 'renamed_file.txt')
print(f"File 'renamed_file.txt' ada: {os.path.exists('renamed_file.txt')}")

# Menghapus file
os.remove('renamed_file.txt')
print(f"File 'renamed_file.txt' ada: {os.path.exists('renamed_file.txt')}")

```

#### 4. Menjalankan Perintah Sistem

Anda dapat menjalankan perintah shell langsung dari Python menggunakan `os.system()` atau `subprocess` module (yang lebih direkomendasikan untuk kasus yang lebih kompleks).

```python
import os

print("Menjalankan perintah 'ls -l' (list files in long format):")
os.system('ls -l')

# Untuk mendapatkan output dari perintah, `subprocess` module lebih disarankan:
# import subprocess
# result = subprocess.run(['ls', '-l'], capture_output=True, text=True)
# print(result.stdout)
```
