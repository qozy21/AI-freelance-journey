##Python standard library : Module Math

Modul `math` di Python menyediakan akses ke fungsi-fungsi matematika yang ditentukan oleh standar C. Modul ini mendukung operasi floating-point standar dan beberapa fungsi yang lebih spesifik.

###Beberapa Fungsi Umum:
* `math.sqrt(x)`: Mengembalikan akar kuadrat dari x.
* `math.pow(x, y)`: Mengembalikan x dipangkatkan y.
* `math.log(x[, base])`: Mengembalikan logaritma natural (basis e) dari x. Jika base ditentukan, mengembalikan logaritma x dengan basis tersebut.
* `math.sin(x)`, `math.cos(x)`, `math.tan(x)`: Fungsi trigonometri untuk sinus, kosinus, dan tangen x (dalam radian).
* `math.radians(x)`: Mengubah sudut x dari derajat ke radian.
* `math.degrees(x)`: Mengubah sudut x dari radian ke derajat.
* `math.floor(x)`: Mengembalikan nilai integer terbesar yang kurang dari atau sama dengan x.
* `math.ceil(x)`: Mengembalikan nilai integer terkecil yang lebih besar dari atau sama dengan x.
* `math.factorial(x)`: Mengembalikan faktorial dari x.
###Beberapa Konstanta Umum:
* `math.pi`: Konstanta matematika pi (Ï€), rasio keliling lingkaran terhadap diameternya.
* `math.e`: Konstanta matematika e, bilangan natural dasar logaritma natural.
* `math.inf`: Infinity (tak terhingga).
* `math.nan`: Not a Number (bukan angka).

```python
import math

#Operasi akar dan pangkat
print(f"Akar kuadrat dari 16: {math.sqrt(16)}")
print(f"2 dipangkat 3: {math.pow(2, 3)}")

#logaritma
print(f"Logaritma natural dari 2: {math.log(2)}")
print(f"Logaritma basis 10 dari 100: {math.log(100, 10)}")

#Pembulatan
print(f"Pembulatan ke bawah (floor) dari 4.7: {math.floor(4.7)}")
print(f"Pembulatan ke atas (ceil) dari 4.3: {math.ceil(4.3)}")
```
