##**Python standard library : Module JSON**

Modul `JSON` adalah format pertukaran data yang ringang dan mudah dibaca oleh manusia serta mudah diurai oleh mesin. format ini sering digunakan untuk mengirim data antara server dan aplikasi web atau untuk menyimpan konfigurasi

**Fungsi Utama**:
1. `json.dumps()`(dump string): mengonversi objek python (dict atau list) menjadi string json, yang berguna untuk menyimpan data json kedalam variable string atau mencetaknya.
2. `json.loads()` (load string): mengonversi string json menjadi objek python. ini sering digunakan ketika menerima data json sebagai string(misalnya dari API) dan ingin memprosesnya dalam python
3. `json.dump()`: mengonversi objek python menjadi string JSON dan menulisnya ke file terbuka. berguna ketika menyimpan data python langsung dalam file `.json`
4. `json.load()`: membaca data json dari file yang terbuka dan mengonverikannya menjadi objek python. digunakan untuk memmuat data dari file`.json` kedalam program

nb: load itu narik data dari json, dan dump itu membuang data ke json


```python
import json

# objek python
data_siswa = {
    "nama": "santoso",
    "umur": 14,
    "keaktifan": True,
    "hobi": ["Bersepeda","Hiking"],
    "alamat": {
        "kota": "semarang",
        "kecamatan": "ungaran"
    }
}

print("--- Objek Python ---")
print(data_siswa)
print(f"Tipe: {type(data_siswa)}\n")
```

```python
# penggunaan json.dumps()
json_string = json.dumps(data_siswa,indent=4) # indent untuk format yang lebih rapi

print("--- JSON String (dari dumps) ---")
print(json_string)
print(f"Tipe: {type(json_string)}\n")
```

```python
# penggunaan json.loads()
data_dari_json_string = json.loads(json_string)

print("--- Objek Python (dari loads) ---")
print(data_dari_json_string)
print(f"Tipe: {type(data_dari_json_string)}\n")
```

```python
# penggunaan json.dump() dan json.load()
file_name = "data_siswa.json"

# Menulis ke file (dump)
with open(file_name, "w") as f:
    json.dump(data_siswa, f, indent=4)
print(f"Data Python berhasil ditulis ke '{file_name}'\n")

# Membaca dari file (load)
with open(file_name, "r") as f:
    data_dari_file = json.load(f)

print("--- Objek Python (dari file load) ---")
print(data_dari_file)
print(f"Tipe: {type(data_dari_file)}\n")
```
